## Import Data

```python
list(sns.get_dataset_names())
```

```python
df_tips = sns.load_dataset("tips")
df_tips.head(5)
```

```python
df_rand = pd.DataFrame(dict(time=np.random.choice(list(range(50)), 50), value=np.random.randn(50).cumsum()))
df_rand.head(4)
```

```python
df_fmri = sns.load_dataset("fmri")
df_fmri.head(4)
```

```python
df_penguins = sns.load_dataset("penguins")
df_penguins.head(4)
```

```python
df_titanic = sns.load_dataset("titanic")
df_titanic.head(4)
```

```python
df_crash = sns.load_dataset("car_crashes")
corr_crash = df_crash.corr()
```

## GrÃ¡fico DispersÃ£o (scatter plot)

#### Style

#### Palette

```python
sns.relplot(x="total_bill", y="tip", size="size", sizes=(15, 200), data=df_tips);
```

#### TendÃªncia - ci

```python
sns.relplot(x="timepoint", y="signal", kind="line", data=df_fmri, ci=None);
```

```python
sns.relplot(x="timepoint", y="signal", kind="line", data=df_fmri);
```

```python
sns.relplot(x="timepoint", y="signal", kind="line", data=df_fmri, ci='sd');
```

#### Estimator

```python
sns.relplot(x="timepoint", y="signal", estimator=None, kind="line", data=df_fmri);
```

#### Hue

```python
sns.relplot(x="timepoint", y="signal", hue="event", kind="line", data=df_fmri);
```

#### Style

```python
sns.relplot(x="timepoint", y="signal", hue="region", style="event", kind="line", data=df_fmri);
```

#### Markers

```python
sns.relplot(x="timepoint", y="signal", hue="region", style="event", dashes=False, markers=True, kind="line", data=df_fmri);
```

#### Hue & Style & ci & Markers

```python
sns.relplot(x="timepoint", y="signal", hue="event", style="event", kind="line", data=df_fmri);
```

```python
sns.relplot(x="timepoint", y="signal", hue="event", style="event", kind="line", ci=None, markers=True, data=df_fmri);
```

#### Units

```python
sns.relplot(x="timepoint", y="signal", hue="region", units="subject", estimator=None, kind="line", data=df_fmri[df_fmri["event"].isin(['stim'])] );
```

```python
sns.relplot(x="timepoint", y="signal", hue="region", style="event", estimator=None, kind="line", data=df_fmri[df_fmri["event"].isin(['stim', 'cue'])] );
```

## Multiplos GrÃ¡ficos

```python
sns.relplot(x="timepoint", y="signal", 
            hue="event", style="event",
            col="subject", col_wrap=5, height=3, 
            aspect=.80, linewidth=2.5, kind="line", 
            data=df_fmri.query("region == 'frontal'"));
```
